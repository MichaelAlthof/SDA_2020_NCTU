```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy

# generate two sets of U(0, 1) random variables
rng = np.random.RandomState(seed=42)
n=10000
u1 = rng.uniform(size=n)
u2 = rng.uniform(size=n)

# generate *independent* normal random variables using Box-Muller
R_sq = -2 * np.log(u1)
theta = 2 * np.pi * u2
z1 = np.sqrt(R_sq) * np.cos(theta)
z2 = np.sqrt(R_sq) * np.sin(theta)

h = sns.jointplot(z1, z2, kind='scatter', marker="+")
h.set_axis_labels('X=r*cos(theta)', 'Y=r*sin(theta)', fontsize=12)

plt.tight_layout()
```
